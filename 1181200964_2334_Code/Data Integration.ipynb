```python
from glob import glob # package
import pandas as pd
import json 
import csv
```

```python
data = glob('Datasets/*') # Read the folder
```

```python
data_list = []

for d in data:
    with open (d) as doc: # Read all the files in the folder
        exp = json.load(doc) # Load the data
        data_list.append(exp) # Data is appended into a data list
```

```python
df = pd.DataFrame(data_list) # Generate dataset with dataframe
```

```python
row = pd.concat([df[['entityId','entityType','event','eventTime']],pd.json_normalize(df['properties'])],axis=1) 
# Select the features in the properties
```

```python
row.to_csv('Dataset.csv', index=False) # Transform dataset into the excel file
```

```python
dfSort = row.sort_values(by='label')
dfSort
```

```python
newDataset = dfSort.fillna("NA")
newDataset
```
