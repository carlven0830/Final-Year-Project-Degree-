```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Dataset(LabelEncoder).csv", low_memory=False)
df
```

```python
X = df.iloc[:, 0:65]
y = df['label']

y.value_counts().plot.pie(autopct='%.2f', ylabel='')
y.value_counts()
```

```python
from imblearn.over_sampling import SMOTE

smt = SMOTE(random_state = 0) # Define SMOTE method
X_sm, y_sm = smt.fit_resample(X, y) # Fit and apply the transform

y_sm.value_counts().plot.pie(autopct='%.2f', ylabel='')
y_sm.value_counts()
```

```python
df2 = pd.DataFrame(X_sm)
df2.to_csv('Dataset(SMOTE).csv', index=False)
```

```python
from imblearn.over_sampling import RandomOverSampler

ros = RandomOverSampler(random_state=0)
X_ros, y_ros = ros.fit_resample(X, y)

y_ros.value_counts()
```

```python
df2 = pd.DataFrame(X_sm)
df2.to_csv('Dataset(ROS).csv', index=False)
```
